```python
def image_extract(categories):
    
    data_       = []
    categories_ = categories
    directory   = "E:\dog_cat_classification"
    
    for i in categories_:
        
        folder_path_ = os.path.join(directory, i)
        
        for image in os.listdir(folder_path_):
            
            label_        = categories_.index(i)
            filepath_     = os.path.join(folder_path_, image)
            imageExtract_ = cv2.imread(filepath_)
            resizeImage_  = cv2.resize(imageExtract_, (150,150))
            data_.append([resizeImage_, label_])
    
    return data_
    
```

```python
categories = ["cat", "dog"]
data_ = image_extract(categories)
```

```python
data_ = random.shuffle(data_)
```

```python
X     = []
y     = []
size_ = 255
for features_, target_ in data_:
    X.append(features_)
    y.append(target_)
X = np.array(X)/255
y = np.array(y)
    
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

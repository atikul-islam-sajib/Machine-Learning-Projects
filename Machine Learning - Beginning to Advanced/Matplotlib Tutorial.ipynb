```python
from matplotlib import pyplot as plt

np.random.seed(12)

X = [float(x) for x in np.linspace(start = 1, stop = 500, num = 40)]
y = [values_ for values_ in np.random.random(40)]
z = [values_ for values_ in np.random.randint(1,100, 40)]

plt.plot(X, '.-')
plt.legend("Straight Line")
plt.plot(y, 'g+--')
plt.legend("Verticle Line")
plt.plot(z, 'r*:')
plt.legend("ZigZag Line")
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
from matplotlib import pyplot as plt

np.random.seed(12)

X = [float(x) for x in np.linspace(start = 1, stop = 500, num = 40)]
y = [values_ for values_ in np.random.random(40)]
z = [values_ for values_ in np.random.randint(1,100, 40)]

plt.plot(X, '.-')
plt.plot(y, 'g+--')
plt.plot(z, 'r*:')
plt.legend(["Straight Line", "Verticle Line","ZigZag Line" ])
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
from matplotlib import pyplot as plt

np.random.seed(12)

X = [float(x) for x in np.linspace(start = 1, stop = 500, num = 40)]
y = [values_ for values_ in np.random.random(40)]
z = [values_ for values_ in np.random.randint(1,100, 40)]

plt.plot(X, '.-', y, 'g+--', z, 'r*:', markersize = 10, alpha = 0.7)
plt.legend(["Straight Line", "Verticle Line","ZigZag Line" ])
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import numpy as np 
import pandas as pd
```

```python
df_ = pd.read_csv('kidney_disease.csv')
df_.head()
```

```python
#find, whether dataset has null value or not
if (df_.isnull().sum().sum())>0:
    print("There is NULL value presence in dataset.")
else:
    print("There is NOT null value in dataset.")
```

```python
#changes the dtypes
#first print the dtypes
for index_, dtypes_ in zip(df_.dtypes.index, df_.dtypes.values):
    print(index_, dtypes_, end = " | ")
print("\n")
#change the dtypes "pcv" and "wc"
list_ = ['pcv', 'wc','rc']
for column_ in list_:
    df_[column_] = pd.to_numeric(df_[column_], errors='coerce')
    print(column_, 'is done.'.title())
print("\n")
#print the data types again
for index_, dtypes_ in zip(df_.dtypes.index, df_.dtypes.values):
    print(index_, dtypes_, end = " | ")
```

```python
#find out which columns have null value and their quantity
nullDataset_ = df_.loc[:, df_.isnull().sum()>0]
#print the null dataset 
nullDataset_.head()
```

```python
#print the index and their quantity
for index_, values_ in zip(nullDataset_.isnull().sum().index, nullDataset_.isnull().sum().values):
    print(index_," ", values_, end = " | ")
print("\n")
#split the dataset with a condition, which have 100+ null value will be fill with "missing", others will be filled with most frequent values
for column_, values_ in zip(nullDataset_.isnull().sum().index, nullDataset_.isnull().sum().values):
    if column_ == 'pcv':
        df_.fillna(df_.loc[:, column_].mean(), inplace = True)
        print(column_, ' is filled with mean.\n')
    elif column_ == 'wc':
        df_.fillna(df_.loc[:, column_].mean(), inplace = True)
        print(column_, ' is filled with mean.\n')
    elif values_>=100:
        df_.loc[:, column_].fillna('missing',  inplace = True)
        print(column_,' is filled with missing.\n')
    else:
        df_.loc[:, column_].fillna(df_.loc[:, column_].value_counts().sort_values(ascending = False).index[0], inplace = True)
        print(column_,' is filled with most frequent approach.\n')
print("Done with null values.\n")
#check whether dataset has still null value or not
if (df_.isnull().sum().sum()) == 0:
    print("There is NO NULL value presence in dataset.")
else:
    print("There is still NULL Values presence in dataset.")
```

```python
#convert the "classification" in a numeric form
for values_, counts_ in zip(df_.classification.value_counts().index, df_.classification.value_counts().values):
    print(values_, 'has ', counts_, 'unique values.')
index_ = df_[df_.classification == 'ckd\t'].index
for dropIndex_ in index_:
    df_.drop([dropIndex_], axis = 0, inplace = True)
    print("Dropped ", dropIndex_,'\n')
print("Done.")
if len(df_.classification.value_counts().index) == 2:
    print("It's binary class classsification problem.\n")
else:
    print("It's categorical cross entropy")
#Now, change this as numeric form
df_.loc[:, 'classification'] = df_.loc[:, 'classification'].map({'ckd': 1, 'notckd': 0})
#print the df_
df_.head(5)
```

```python
#first convert 
objectColumns_ = df_.select_dtypes(include = 'object').columns
#print the columns name
for column_ in objectColumns_:
    print(column_, end = " | ")
print("\n")
#check the unique values
for column_ in objectColumns_:
    print(df_.loc[:, column_].nunique(), end = " | ")
print("\n")
#Now delete the "id" and "classification" columns and perform get dummies
target_ = df_.loc[:, 'classification']
delete_ = ['id', 'classification']
for delColumn_ in delete_:
    df_.drop([delColumn_], axis = 1, inplace = True)
    print(delColumn_, ' is deleted. ')
#now, use "get_dummies"
getDummies_ = pd.get_dummies(df_[objectColumns_], drop_first = True)
#now, we will have to delete the preious column, which were not in numeric
for dropped_ in objectColumns_:
    df_.drop([dropped_], axis = 1, inplace = True)
    print("Column ", dropped_, ' is dropped.')
#now store this in the main data frame
df_ = pd.concat([df_, getDummies_], axis = 1)
#print the df_
df_.head()
```
